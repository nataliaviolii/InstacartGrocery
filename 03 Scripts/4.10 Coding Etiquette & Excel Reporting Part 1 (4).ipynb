```python
# Create an exclusion flag for low-activity customers 

df_project_all.loc[df_project_all['max_order'] < 5, 'order_activity_flag'] = 'Low order activity customer'
df_project_all.loc[df_project_all['max_order'] >= 5, 'order_activity_flag'] = 'High order activity customer'
```

```python
df_project_all['order_activity_flag'].value_counts()
```

```python
# Create a sample that excludes all low activity customers

df_exc_low_cust = df_project_all[df_project_all['order_activity_flag'] == 'High order activity customer']
```
