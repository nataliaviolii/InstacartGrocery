```python
df_dep_t
```

```python
# Your actual headers are located in the first row of your dataframe: "department_id" and "department."
# How can you get rid of the “0” and turn the first row of your dataframe into your headers?
# Let’s start by adding something to your dataframe: an index. 
# You can add an index to a dataframe that doesn’t have one with the following function:
df_dep_t.reset_index()
```

#### Creating a new header

```python
# This will be done in three steps:
# 1) Create a new header; 
# 2) Remove the first row in the dataframe;
# 3) Add the new header.
```

```python
# First, you need to tell Python to take the row under the current header (the first row in the dataframe),
# use that data to create a new header, and get rid of the current header. 

# Remember, Python starts indexing at 0 instead of 1, so the first row would be row 0.
# Start by creating an object that contains the first row of the dataframe, row 0, using the following command:

# Take the first row of df_dep_t for the header
new_header = df_dep_t.iloc[0]

# The iloc[] function is used for selecting data in pandas and takes the index number of a row or column (or multiple rows and columns).
# You can think of it using a little mnemonic: “locate an integer” = iloc.
```

```python
# By passing row 0 into this function and setting it to a new variable, new_header,
# you’re effectively copying the values within row 0 of your dataframe and assigning them to this variable for easy access. 
new_header
```

```python
# let’s create a new dataframe that only copies over rows beyond the first row.
# What does that 1: mean? This is telling Python to copy everything from the first row onward from the df_dep_t dataframe into a new dataframe, df_dep_t_new.
# The 1 represents row 1 in the dataframe.
# Remember that since Python starts indexing from 0 rather than 1, “row 1” is actually the second row in the dataframe.
# The colon that follows the 1 represents onward until the end of the dataframe.
# Together, they tell Python to copy everything starting from row 1 until the end of the dataframe.

df_dep_t_new = df_dep_t[1:]
```

```python
# Try calling your new dataframe

df_dep_t_new
```

```python
# All that’s left to do now is set your new_header variable as the new header for your df_dep_t_new dataframe.
# To that end, all you need to do is tell Python to use the list of column names in your new_header variable as your new header. I

df_dep_t_new.columns = new_header
```

```python
# You probably remember the df.columns function from the previous Exercise, when you used it to show the names of the columns in a dataframe.
# You can also use df.columns to assign names to the columns in a dataframe.
# The command above says “take the values stored in the new_header variable and put them in the df_dep_t_new dataframe as the names of its columns.”

# Now let see the result
df_dep_t_new
```

### Subsetting

```python
# Subsetting refers to creating a smaller data set from a whole data set (in this case, df_prods) based on a particular filter
```

```python
# Let’s create a subset for your df_prods dataframe that only contains data from the snacks department. 

df_snacks =  df_prods[df_prods['department_id'] == 19]

# Within the brackets on the right-hand side, you’re telling Python to look into the df_prods dataframe
# and find a column called "department_id."
# And, within that "department_id" column, it should look for a value of 19.

# The object['element'] syntax is called indexing. 
# Indexing refers to retrieving some element from an object, be it a dataframe, list, or otherwise. 
# By using the index operator,[], you’re telling Python to look for the element in the brackets within the object to the left of the brackets
# (so, in this case, looking for a "department_id" value of 19 within the df_prods dataframe).
```

```python
df_prods['department_id'] == 19

# The indexing code on its own simply searches for the data in question:
# in this case, everything with a "department_id" of 19 within the df_prods dataframe.
# The output, here, shows you how Python looks for this data.
# You can see it traveling through every row within the df_prods dataframe and assigning each value either True or False depending on whether it meets the criteria. 
```

```python
df_snacks =  df_prods[df_prods['department_id'] == 19]

# Enclosing df_prods['department_id'] == 19 inside df_prods[] tells Python to map the true/false procedure you witnessed in the previous output onto the columns within the df_prods dataframe
# the result being a list of only those values within df_prods that are true
# (and, subsequently, only those values within df_prods that have a "department_id" of 19).
```

```python
df_snacks
```

```python
df_snacks.head()
```

```python
# There’s a second way of achieving these same results that you might find a bit simpler. This involves use of the loc function.
# The loc function is similar to the iloc function you used previously,
# but while iloc looks for an integer position, loc looks for a column name.
# In your case, you want this function to look for instances where the value of "department_id" is 19 within the df_prods database:

df_snacks_2 = df_prods.loc[df_prods['department_id'] == 19]

# You could also use the command:
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]

# In this second version, rather than using an equals sign, you’re telling the loc function to look into a list: isin([19]).
# This is handy because you could use more than one value here if you wanted to check multiple "department_ids" at the same time (isin([17,18,19])).
```
